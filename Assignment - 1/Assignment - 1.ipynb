```python
Q1. Explain with an example each when to use a for loop and a while loop.

Ans. 
=> When starting index and ending index both are known then for loop is used.
For example : Print 1 to 10
    for i in range(10):
        print(i)
        
=> while loop is used when we have one condition.
For example : print i while i is less than 10
    i = 1
    while i < 10:
        print(i)
        i++
```

```python
Q2. Write a python program to print the sum and product of the first 10 natural numbers using for
and while loop.

Ans.
=> Using for loop

sum = 0
product = 1

for i in range(10):
    sum += i+1
    product *= i+1
print("Sum = " ,sum)
print("Product = ",product)

=> Using while loop

sum = 0
product = 1
i = 1

while i <= 10 :
    sum += i
    product *= i
    
    i+=1

print("Sum = " ,sum)
print("Product = ",product)
```

```python
Q3. Create a python program to compute the electricity bill for a household.
The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.

0 to 100 -> 4.5
101 to 200 -> 6
201 to 300 -> 10
>300 -> 20

Ans.

units = int(input("How much units are consumed"))
bill = 0

if(units > 100):
    bill += 450
    units -= 100
    
    if(units > 100):
        bill += 600
        units -= 100
        
        if(units > 100):
            bill += 1000
            units -= 100
            
            bill += units * 20
        else:
            bill += units * 10
    else:
        bill += units * 6
else:
    bill = units * 4.5
    
print("Total electicity bill is ",bill)
```

```python
Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
that list.

Ans.

=> Using for loop

main_list = list(range(1,101))
ans_list = []

for i in main_list:
    if((i*i*i) % 4 == 0 or (i*i*i) % 5 == 0):
        ans_list.append(i)
        
print(ans_list)


=> Using while loop

main_list = list(range(1,101))
ans_list = []
i = 0

while i < 100:
    if(((main_list[i]*main_list[i]*main_list[i]) % 4 == 0) or ((main_list[i]*main_list[i]*main_list[i]) % 5 == 0)):
        ans_list.append(main_list[i])
        
    i += 1
        
print(ans_list)
```

```python
Q5. Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"

Ans.

string = "I want to become a data scientist"
string = string.upper()

count = 0

for i in string:
    if(i == 'A' or i == 'E' or i == 'I' or i == 'O' or i == 'U'):
        count += 1
        
print("Count of vowels is ",count)
```
