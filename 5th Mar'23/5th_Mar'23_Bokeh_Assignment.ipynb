```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

# Ans.

# glyphs are visual markers used to represent data on a plot

from bokeh.plotting import figure, show

p = figure(width=500, height=400, title='Scatter Plot')

p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=12, color='blue')
p.square([2, 3, 4, 5, 6], [5, 4, 3, 2, 1], size=10, color='red')

show(p)
```
