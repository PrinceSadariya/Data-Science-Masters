```python
# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# Ans.

# Scatter Plot: A scatter plot shows the relationship between two numerical variables, with each data point represented as a dot on the plot. It is useful for visualizing the correlation or pattern between two variables.

# Bar Plot: A bar plot displays categorical data using rectangular bars, with the height of each bar representing a specific value. It is commonly used to compare values across different categories or to show the distribution of a categorical variable.

# Histogram: A histogram represents the distribution of a continuous numerical variable by dividing it into bins and displaying the frequency or count of observations within each bin. It helps in understanding the underlying distribution of the data.

# Box Plot: A box plot (or box-and-whisker plot) visualizes the distribution of a numerical variable across different categories. It provides information about the median, quartiles, and potential outliers, making it useful for identifying differences or similarities between groups.

# Heatmap: A heatmap is a graphical representation of data in a matrix format, where each cell's color represents the value of the corresponding variable. It is often used to visualize correlations in a dataset or to highlight patterns and relationships.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.

# Ans.

import seaborn as sns

fmri = sns.load_dataset("fmri")

sns.lineplot(x=fmri.timepoint,y=fmri.signal,data=fmri,hue="event",style="region")
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.

# Ans.

titanic = sns.load_dataset("titanic")

sns.boxplot(x=titanic.pclass,y=titanic.fare)
sns.boxplot(x=titanic.pclass,y=titanic.age)
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.

# Ans.

diamonds = sns.load_dataset("diamonds")
sns.histplot(x=diamonds.price,data=diamonds,hue="cut")
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.

# Ans.

iris = sns.load_dataset("iris")

sns.pairplot(iris,hue="species")
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

# Ans.

flights = sns.load_dataset("flights")

sns.heatmap(data = flights[['year','passengers']])
```
