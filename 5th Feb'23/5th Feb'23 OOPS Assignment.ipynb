```python
# Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

# Ans.

# => In OOP class is a blueprint for creating object. In which all methods and properties are defined which you can use
# => After declartion of class we can create multiple object from the class
# => Object is a instance of the class

# Example

# Here we are taking example of person
# Person can be male or female
# Person have his/her name,age etc.

class Person:
    def __init__(self,name,gender,age):
        self.name = name
        self.gender = gender
        self.age = age
        
    def intro(self):
        print(f"{self.name} is {self.age} years old")
        
nobita = Person("Nobita","Male",20)
sizuka = Person("Sizuka","Female",19)

nobita.intro()
sizuka.intro()
```

```python
# Q5. What is inheritance? Give an example for each type of inheritance.

# Ans.

# When we need properties of particular claas into another class then inheritance is used
# In inheritance child class aquire all propertirs/methods of parent calss

# There are 5 types of inheritance
# (1) Single Inheritance
# (2) Multilevel Inheritance
# (3) Multiple Inheritance
# (4) Hierarchical Inheritance
# (5) Hybrid Inheritance
```

```python
# (1) Single Inheritance

class Person:
    def __init__(self,name,gender,age):
        self.name = name
        self.gender = gender
        self.age = age
        
class Employee(Person):
    
    def set_details(self,e_id,department,role):
        self.e_id = e_id
        self.department = department
        self.role = role
        
    def get_details(self):
        print("Employee Name : ",self.name)
        print("Employee Gender : ",self.gender)
        print("Employee Age : ",self.age)
        print("Employee ID : ",self.e_id)
        print("Employee Department : ",self.department)
        print("Employee Role : ",self.role)
        
emp1 = Employee("Nobita","Male",20)

emp1.set_details(1111,"Data Science"," Sr. Data Engineer")

emp1.get_details()
```

```python
# (2) Multilevel Inheritance

class c1:
    def c1_fun(self):
        print("This is class 1")
        
class c2(c1):
    def c2_fun(self):
        print("This is class 2")

class c3(c2):
    def c3_fun(self):
        print("This is class 3")
        
obj = c3()
obj.c1_fun()
obj.c2_fun()
obj.c3_fun()

# here we can use c1'methods and c2's methods from c3
```

```python
# (3) Multiple Inheritance

class M1:
    def m1_fun(self):
        print("This is m1 class")
        
class M2:
    def m2_fun(self):
        print("This is m2 class")
        
# This is main claas (M) which is inherit both m1 and m2

class M(M1,M2):
    def m_fun(self):
        print("This is main M class")
        
obj = M()
obj.m1_fun()
obj.m2_fun()
obj.m_fun()
```

```python
# (4) Hierarchical Inheritance

# When more than one derived class are created from a single base this type of inheritance is called hierarchical inheritance.

class P:
    def p_fun(self):
        print("This is P class")
        
class C1(P):
    def c1_fun(self):
        print("This is C1 Class")
        
class C2(P):
    def c2_fun(self):
        print("This is C2 Class")
        
class C3(P):
    def c3_fun(self):
        print("This is C3 Class")
```

```python
# (5) Hybrid Inheritance

# Inheritance consisting of multiple types of inheritance is called hybrid inheritance.

class P1:
    def p1_fun(self):
        print("This is P1 class")
        
class P2:
    def p2_fun(self):
        print("This is P2 class")
        
class C1(P1):
    def c1_fun(self):
        print("This is C1 Class")
        
class C2(P2,C1):
    def c2_fun(self):
        print("This is C2 Class")
```
