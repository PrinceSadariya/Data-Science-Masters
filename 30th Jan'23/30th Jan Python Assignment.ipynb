```python
# Q.1 Write a program to accept percentage from the user and display the grade according to the following criteria:
    
#     >90 -> A
#     >80 and <=90 -> B
#     >= 60 and <=80 -> C
#     bellow 60 -> D

# Ans.

pr = int(input("Enter your percentage"))
if pr > 90 and pr <= 100:
    print("A Grade")
elif pr > 80 and pr <= 90 :
    print("B Grade")
elif pr >= 60 and pr <= 80 :
    print("C Grade")
elif pr < 60 and pr >= 0 :
    print("D Grade")
else:
    print("Please enter valid percentage")
```

```python
# Q.2 Write a program to accept the cost price of a bike and display the road tax to be paid according to the following criteria:
    
#     15% > 100000
#     10% > 50000 and <=100000
#     5% <= 50000

price = int(input("Enter bike price"))

if price > 100000 :
    print("Road Tax to be paid is : ",(price*15)/100)
elif price >50000 and price <= 100000 :
    print("Road Tax to be paid is : ",(price*10)/100)
elif price >=0 and price <= 50000 :
    print("Road Tax to be paid is : ",(price*5)/100)
else :
    print("Please enter valid input")
```

```python
# Q.3 Accept any city from the user and display monuments of that city.

# Ans.

print("Press 1 for Delhi , 2 for Agra , 3 for Jaipur")

city = int(input("Enter number here"))

match city:
    case 1:
        print("Red Fort")
    case 2:
        print("Taj Mahal")
    case 3:
        print("Jal Mahal")
    case _:
        print("Please enter valid input")
```

```python
# Q.4 Check how many times a given numer can be divided by 3 before it is less than or equel to 10

# Ans.

# Approch 1
num = int(input("Enter number "))
cnt = 0

while(num >= 10):
    if num % 3 == 0:
        cnt += 1
        num /= 3
    else :
        break
        
print("Given number can be divded", cnt ,"times ,before it is less than or equel to 10")


# Approch 2
num = float(input("Enter number "))
cnt = 0

while(num >= 10):
    num /= 3
    cnt += 1
        
print("Given number can be divded", cnt ,"times ,before it is less than or equel to 10")
```

```python
# Q.6 Use nested while loop to print 3 different pattern.

# Ans.

# Pattern 1

n = int(input("Enter number of row"))
i = 0
j = 0

while i<n:
    while j<=i:
        print("*",end=" ")
        j += 1
    print("")
    i += 1
    j = 0
    
# Pattern 2

n = int(input("Enter number of row"))
i = 0
j = 0

while i<n:
    while j<n-i:
        print("*",end=" ")
        j += 1
    print()
    i += 1
    j = 0
    
# Pattern 3

n = int(input("Enter number of row"))
i = 0
j = 0

while i<n:
    while j<n:
        if i==0 or i==n-1 or j==0 or j==n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
        j += 1
    print("")
    i += 1
    j = 0
```
