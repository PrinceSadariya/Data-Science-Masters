```python
# Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in
# predicting the quality of wine.

# Ans.

# Fixed Acidity: This feature represents the non-volatile acids present in the wine, which contribute to the overall acidity level. Acidity is an essential characteristic that affects the taste and balance of the wine.

# Volatile Acidity: Volatile acids are responsible for the unpleasant vinegar-like aroma in wine. High levels of volatile acidity can negatively impact the quality and taste of wine.

# Citric Acid: Citric acid contributes to the freshness and tartness of the wine. It can enhance the flavor and provide a balancing effect, particularly in white wines.

# Residual Sugar: This feature represents the amount of sugar that remains in the wine after fermentation. It influences the perceived sweetness of the wine and can affect its overall balance.

# Chlorides: Chlorides in wine can come from various sources, including the soil and winemaking practices. They can impact the wine's taste and contribute to its salinity or saltiness.

# Free Sulfur Dioxide: Sulfur dioxide is commonly used in winemaking as a preservative and antioxidant. The free form of sulfur dioxide is responsible for protecting the wine from oxidation and microbial spoilage.

# Total Sulfur Dioxide: This feature represents the total amount of both free and bound forms of sulfur dioxide in the wine. It is an important indicator of the wine's stability and preservation.

# Density: Density refers to the mass of the wine per unit volume. It can provide insights into the sugar and alcohol content of the wine.

# pH: pH is a measure of the acidity or alkalinity of the wine. It plays a crucial role in the wine's overall balance, taste, and microbial stability.

# Sulphates: Sulphates are a type of salt that can be found in wine. They are believed to contribute to the wine's antioxidant properties and can affect its taste and longevity.

# Alcohol: The alcohol content of wine influences its body, mouthfeel, and perceived sweetness. It can significantly impact the wine's overall quality and consumer preference.

# Quality: This is the target variable in the dataset, representing the perceived quality or rating of the wine. It is often assessed by wine experts or through sensory evaluations.
```

```python
# Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
# features. What is the minimum number of principal components required to explain 90% of the variance in
# the data?

# Ans.

import numpy as np
from sklearn.decomposition import PCA

df_std = (df - df.mean()) / df.std()

pca = PCA()
pca.fit(df_std)

explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

n_components = np.argmax(cumulative_variance_ratio >= 0.9) + 1

print("Explained Variance Ratio:", explained_variance_ratio)
print("Cumulative Variance Ratio:", cumulative_variance_ratio)
print("Number of Principal Components for 90% variance:", n_components)
```
