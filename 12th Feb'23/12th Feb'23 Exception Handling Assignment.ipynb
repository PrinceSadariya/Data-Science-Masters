```python
# Q2. What happens when an exception is not handled? Explain with an example.

# Ans.

# If exception is not handled , then program excecution is stop , and remaining line of code will not execute

print("Hello program start")

f = open("temp.txt",'r')

print("This is end of program")

# here last line "This is end of program" will not excecute
```

```python
# Q3. Which Python statements are used to catch and handle exceptions? Explain with an example

# Ans.

# Normally try and except statement is used to catch and handle exceptions

print("Program start")

try:
    f = open("temp.txt",'r')
except Exception as e:
    print("Error in file opening " , e)
    
print("Program end")
```

```python
# Q.4 Explain with an examole
#     a. try and else
#     b.finally
#     c.raise

# a.try and else

# in try block we have to run our code , which might be thrwo an error
# if code of try block execute without an error , then code of else block will run 

print("Program start")

try:
    print("This is code of try block")

except Exception as e:
    print("Error in try block " , e)
    
else:
    print("This is code of else block")
    
print("Program End")
```

```python
# b. finally

# no matter code of try block throw an error or not , code of finally block will excecute

print("Program start")

try:
    f - open('temp.txt','r')

except Exception as e:
    print("Error in try block " , e)
    
else:
    print("This is code of else block")
    
finally:
    print("This is code of finally block")
    
print("Program End")
```

```python
# c. raise

# raise is used to handle custom exception

class ValidateAge(Exception):
    def __init__(self,msg):
        self.msg = msg
        
def validAge(age):
    if type(age) is int:
        if age < 0:
            raise ValidateAge("Age cant be less than zero")
        elif age > 150:
            raise ValidateAge("Entered age is very high")
        else:
            print("Age is valid")
    else:
        raise ValidateAge("Please enter valid input")
        
try :
    age = int(input("Enter ypu age"))
    validAge(age)
except Exception as e:
    print(e)
```

```python
# Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example

# Ans.

# witch exceptions is defined by user(developer) is known as custom exceptions
# python provide many buit-in exceptions , but sometime we have to implement our conditions to check
# so in that case we can not use built-in exceptions , we have to define our Custom exception

# suppose here we have to check entered age is grater than zero and less than 150 , and age can not be string , then we have to make custom exception

class ValidateAge(Exception):
    def __init__(self,msg):
        self.msg = msg
        
def validAge(age):
    if type(age) is int:
        if age < 0:
            raise ValidateAge("Age cant be less than zero")
        elif age > 150:
            raise ValidateAge("Entered age is very high")
        else:
            print("Age is valid")
    else:
        raise ValidateAge("Please enter valid input")
        
try :
    age = int(input("Enter ypu age"))
    validAge(age)
except Exception as e:
    print(e)
```
